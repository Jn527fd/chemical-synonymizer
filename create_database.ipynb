PubChemLite Database

```python
PubChemLite = pd.read_csv("data/PubChemLite.csv")
print("Rows:", len(PubChemLite))
print("Columns:", PubChemLite.columns)
display(PubChemLite.head())
```

```python
def smart_split(s):
    return re.split(r";(?![^()]*\))", s)

syn_counts = (
    PubChemLite["Synonym"]
    .fillna("")
    .astype(str)
    .apply(lambda x: len([s for s in smart_split(x) if s.strip()]))
)

num_multi_name = (syn_counts > 1).sum()
print("Molecules with multiple names:", num_multi_name)

PubChemLite_multiple_synonyms = PubChemLite[syn_counts > 1]
display(PubChemLite_multiple_synonyms)

rows = []
for _, row in PubChemLite_multiple_synonyms.iterrows():
    cid, synonyms = row["Identifier"], str(row["Synonym"]).split(";")
    
    for name in synonyms:
        name = name.strip().lower()
        if name:
            rows.append((name, cid))

PubChemLite_synonyms = pd.DataFrame(rows, columns=["name", "cid"])
print("Total names (only multi-synonym molecules):", len(PubChemLite_multiple_synonyms))
PubChemLite_synonyms.head()
```

which names are already present in ChEBI and do not need to be extracted from PubChemLite

```python
PubChemLite_synonyms["name_clean"] = PubChemLite_synonyms["name"].str.lower().str.strip()
ChEBI["ascii_name_clean"] = ChEBI["ascii_name"].str.lower().str.strip()
```

```python
matches = PubChemLite_synonyms.merge(
    ChEBI[["compound_id", "ascii_name_clean"]],
    left_on="name_clean",
    right_on="ascii_name_clean",
    how="left"
)

display(matches)
```

```python
display(PubChemLite_multiple_synonyms[PubChemLite_multiple_synonyms["Identifier"] == 20652])
display(PubChemLite_multiple_synonyms[PubChemLite_multiple_synonyms["Identifier"] == 21639])
```

```python
found = matches[matches["compound_id"].notna()].drop_duplicates(subset=["name", "cid"])
missing = matches[matches["compound_id"].isna()].drop_duplicates(subset=["name", "cid"])
display(found)
display(missing)
```

```python
cid_to_compound = dict(
    found[["cid", "compound_id"]].drop_duplicates().values
)

missing["compound_id"] = missing["cid"].map(cid_to_compound)
```

```python
ChEBI_useful_data = ChEBI[["compound_id", "ascii_name_clean"]].dropna()
Extra_data = missing_not_nan[["compound_id", "ascii_name_clean"]].dropna()
combined = pd.concat([ChEBI_useful_data, Extra_data], ignore_index=True)
```
